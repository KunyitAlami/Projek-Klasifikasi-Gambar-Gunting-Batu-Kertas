```python
import tensorflow as tf
import keras_preprocessing
from keras_preprocessing import image
from keras_preprocessing.image import ImageDataGenerator
```

Kita akan membuat folder awal untuk mengakses dataset dari gambar gunting batu kertas

```python
folder_dasar = "/content/rockpaperscissors/rps-cv-images"
```

Setelah membuat folder dasar kita akan melakukan augmentasi dan melakukan split  data untuk data training dan data validation

```python
pelatihan_datagen = ImageDataGenerator(
      rescale = 1./255,
#kita akan melakukan syarat-syarat augmentasi
	    rotation_range=20,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest',
#kita akan pecah data nya dengan membagi 2
      validation_split=0.4
#berarti 60% data untuk training, dan 40% data untuk validation
      )
```

```python
generator_pelatihan = pelatihan_datagen.flow_from_directory(folder_dasar, target_size=(150,150), class_mode='categorical',
                                                            batch_size=40,shuffle=True,subset='training')

generator_validasi = pelatihan_datagen.flow_from_directory(folder_dasar, target_size=(150,150), class_mode='categorical',
                                                           batch_size=40,shuffle=True, subset='validation')
```

Kita akan membuat model saraf tiruan

```python
model = tf.keras.models.Sequential([

    tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.4),

    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.4),

    tf.keras.layers.Flatten(),


    tf.keras.layers.Dense(521, activation='relu'),

    tf.keras.layers.Dense(256, activation='relu'),

    tf.keras.layers.Dense(3, activation='softmax')
])

model.summary()
```

Kita akan mengomplikasi model

```python
model.compile(loss = 'categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
#kita akan membuat metode callback
class penghentian_pembelajaran_mesin(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.96):
      print(f"\nakurasi lebih dari 0.96")
      self.model.stop_training=True

callbacks=penghentian_pembelajaran_mesin()

komplikasi_model=model.fit(
    generator_pelatihan,
    batch_size=15,
    epochs=10,
    validation_data=generator_validasi,
    validation_steps=5,
    callbacks = [callbacks]
    )
```

```python
from google.colab import files
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input

unggah_gambar = files.upload()

for nama_awal in unggah_gambar.keys():

  jalur = nama_awal
  gambar_yang_akan_diproses = image.load_img(jalur, target_size=(150,150))
  tampilkan_gambar = plt.imshow(gambar_yang_akan_diproses)

  nilai_array_gambar = image.img_to_array(gambar_yang_akan_diproses)
  nilai_array_gambar = np.expand_dims(nilai_array_gambar, axis=0)


  gabung_nilai_array = np.vstack([nilai_array_gambar])
  kelas_gambar = model.predict(gabung_nilai_array, batch_size=10)
  prediksi_kelas_gambar = np.argmax(kelas_gambar)

  print(nama_awal)
  if prediksi_kelas_gambar == 0:
    print('Gambar Ini adalah Paper atau Kertas')
  elif prediksi_kelas_gambar== 1:
    print('Gambar Ini adalah Rock atau Batu')
  else:
    print('Gambar Ini adalah Scissors atau Gunting')
```
